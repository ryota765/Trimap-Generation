```python
import matplotlib.pyplot as plt
import numpy as np
import skimage.segmentation
from PIL import Image
```

```python
# read sample image
img = np.array(Image.open('image/post.jpg'))
plt.imshow(img)
```

## Over-segmentation
###### Usage of SLIC segmentation (https://ieeexplore.ieee.org/document/6205760)

```python
segments_number = 100
max_iterations = 10

segmentation = skimage.segmentation.slic(img,n_segments=segments_number, max_iter=max_iterations)

plt.subplot(1,2,1)
plt.imshow(segmentation)
plt.subplot(1,2,2)
plt.imshow(skimage.segmentation.mark_boundaries(img, segmentation))
```
